```python
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image

```

```python
img = cv.imread("00.jpg")
```

```python
# 3D rgb plot of the original image
image_path = "00.jpg"

fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d") # 3D plot with scalar values in each axis

im = Image.open(image_path)
r, g, b = list(im.getdata(0)), list(im.getdata(1)), list(im.getdata(2))

axis.scatter(r, g, b, c="#ff0000", marker="o")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
# Normalize the image
normalized_img = img/255
cv.imwrite('normalized_img.jpg', normalized_img)
# cv.imshow('test Image', test)
# cv.waitKey(5000)
# cv.destroyAllWindows()
```

```python
# 3D rgb plot of the normalized image

image_path = "normalized_img.jpg"

fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d") # 3D plot with scalar values in each axis

im = Image.open(image_path)
r, g, b = list(im.getdata(0)), list(im.getdata(1)), list(im.getdata(2))

axis.scatter(r, g, b, c="#54FF33", marker="o")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")

axis.axes.set_xlim3d(left=-10, right=10) 
axis.axes.set_ylim3d(bottom=-10, top=10) 
axis.axes.set_zlim3d(bottom=-10, top=10) 

plt.show()
```
