### Cuboids

```python
import gzip
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

# **Folders with more pedestrians**

```python
#Load the first ten cuboids datasets(00 to 09)
label1 = pd.DataFrame()
for i in range(10):
    with gzip.open('cuboids1/cuboids/'+'0'+str(i)+'.pkl.gz', 'rb') as annotated_data:
        data = pickle.load(annotated_data)
        data = pd.DataFrame(data)
        label1 = pd.concat([data, label1])
        
# Load the rest cuboids datasets
for i in range(10,80):
    with gzip.open('cuboids1/cuboids/'+ str(i)+'.pkl.gz', 'rb') as annotated_data:
        data = pickle.load(annotated_data)
        data = pd.DataFrame(data)
        label1 = pd.concat([data, label1])
```

```python
# Cuboids dataset
plot_data1 = pd.DataFrame(label1)
plot_data1.head()
```

```python
# Plot the cuboid annotation distribution based on the labelled column
plot = plot_data1['label'].value_counts().sort_values()
plt.figure(figsize=(15,7))

ax = sns.countplot(x="label", data=plot_data1, order=plot_data1['label'].value_counts().index)
for p in ax.patches:
   ax.annotate('{:d}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()))
plt.xticks(rotation=90)
plt.title('Cuboid annotations')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

# **Semantic Segmentation**

```python
# Load the first ten segmentation datasets(00 to 09)
classes = pd.DataFrame()
for i in range(10):
    with gzip.open('semseg/'+'0'+str(i)+'.pkl.gz', 'rb') as annotated_data:
        data = pickle.load(annotated_data)
        data = pd.DataFrame(data)
        classes = pd.concat([data, classes])
```

```python
# Load the rest segmentation datasets
for i in range(10,80):
    with gzip.open('semseg/'+ str(i)+'.pkl.gz', 'rb') as annotated_data:
        data = pickle.load(annotated_data)
        data = pd.DataFrame(data)
        classes = pd.concat([data, classes])
```

```python
# Segmentation datasets
plot_classes = pd.DataFrame(classes)
plot_classes.head()
```

```python
# Plot the semantic segmentation distribution based on classes

plot = plot_classes['class'].value_counts().sort_values()
plt.figure(figsize=(20,7))

ax = sns.countplot(x="class", data=plot_classes, order=plot_classes['class'].value_counts().index)
for p in ax.patches:
   ax.annotate('{:d}'.format(p.get_height()), (p.get_x(), p.get_height()))
plt.title('Semantic Segmentation')
plt.xlabel('Classes')
plt.ylabel('Count')
plt.show()
```
